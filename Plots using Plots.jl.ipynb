# Data Visualization with Julia
The aim of this notebook is to learn how to visualize data using the `Plots` library in Julia.

```julia
using Pkg

Pkg.add("Plots")

using Plots
using DataFrames
```

Let's start plotting some simple trigonometric functions!

```julia
x = range(0, 10, length=100)
y = sin.(x)
plot(x, y)
```

In `Plots.jl`, every column is a series, that is the set of related points forming the nice line displayed above.

Understanding that allows us to display multiple lines using Matrices:

```julia
x = range(0, 10, length=100)
y1 = sin.(x)
y2 = cos.(x)

plot(x, [y1 y2])
```

We can keep adding lines or series by mutating the plot object.
This time we'll be using the `@.` macro, which inserts points between every function in order to broadcast their operations:

```julia
y3 = @. sin(x)^2 - 1/2    # Equivalent to `y3 = sin.(x).^2.-1/2`

# We mutate the previous plot using `plot!()`
plot!(x, y3)
```

We could have achieved the same by storing the plot in the variable `p` and using it as the first argument of `plot!()`:

```julia
x = range(0, 10, length=100)
y1 = sin.(x)
y2 = cos.(x)
p = plot(x, [y1 y2])

y3 = @. sin(x)^2 - 1/2
plot!(p, x, y3)
```

We can save this figures using different functions depending on the file format we want:
- `savefig()`
- `png()`
- `Plots.pdf()`

Let's style our plot!

```julia
x = range(0, 10, length=100)
y1 = sin.(x)
y2 = cos.(x)

plot(x, [y1 y2],
    title="Trigonometric Functions",
    label=["sin(x)" "cos(x)"],
    linewidth=3
)
```
