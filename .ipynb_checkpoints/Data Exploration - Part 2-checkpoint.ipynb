```julia
using Pkg; Pkg.activate("./")
using StatsPlots
using CSV
using DataFrames
```

```julia
df = CSV.read("relevant_categories.csv", DataFrame)
first(df, 5)
```

```julia
using Random

qty = Int(round(nrow(df)*0.6))

selectedrows = shuffle(1:nrow(df))[1:qty]

sample_set = df[selectedrows, :]

first(sample_set, 5)
```

```julia
numeric_mask = [t == Int64 || t == Float64 for t in eltype.(eachcol(sample_set))]
numeric_sample = sample_set[:, numeric_mask]
first(numeric_sample, 5)
```

```julia
newnames = [replace(n, "-" => "_") for n in names(numeric_sample)]

rename!(numeric_sample, newnames)
```

```julia
gr(size=(800, 600))
@df numeric_sample corrplot([:symboling :normalized_losses :wheel_base :price], fillcolor=cgrad())
```

The biggest correlation with price here is the wheel base of the cars. Both the symboling and normalized losses are slightly correlated with the price too... 

```julia
@df numeric_sample corrplot([:length :width :height :price], fillcolor=cgrad())
```

Both the length and width of the cars are highly correlated with the price, height shows a slight correlation though.

```julia
    @df numeric_sample corrplot([:curb_weight :engine_size :bore :price], fillcolor=cgrad())
```

The last row of plots says it all...

```julia
@df numeric_sample corrplot([:stroke :compression_ratio :horsepower :price], fillcolor=cgrad())
```

This is a little bit more interesting: Among those three features, only the horsepower has a relevant correlation with price.

```julia
@df numeric_sample corrplot([:peak_rpm :city_mpg :highway_mpg :price], fillcolor=cgrad())
```

Ok, with this information we can select the most appropiate set of features!

```julia
dropped_features = ["stroke", "compression-ratio", "peak-rpm"]
```

```julia
final_selection = select(df, Not(dropped_features))
first(final_selection, 5)
```

```julia
CSV.write("selected_features.csv", final_selection)
```
