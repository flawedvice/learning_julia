```julia
using Pkg; Pkg.add("StatsPlots"); Pkg.add("RDatasets")

using DataFrames
using StatsPlots
using RDatasets
```

```julia
# Get iris dataset:
iris = dataset("datasets", "iris")

palette(:default)

# Create a correlational matrix
gr(size = (600, 500))
@df iris corrplot([:SepalLength :SepalWidth :PetalLength :PetalWidth], fillcolor=cgrad())
```

```julia
M = randn(1000,4)
M[:,2] .+= 0.8sqrt.(abs.(M[:,1])) .- 0.5M[:,3] .+ 5
M[:,3] .-= 0.7M[:,1].^2 .+ 2
corrplot(M, label = ["x$i" for i=1:4])
```

```julia
function f(x, y)
    r = sqrt(x^2 + y^2)
    return cos(r) / (1 + r)
end
x = range(0, 2, length = 30)
heatmap(x, x, f, c = :thermal)
```
